# Preparing the data 

```python
# answer= answers['answer']
# text=df['text']
def get_index(text,short_answers):
    index=[]
    for i in df['Unnamed: 0']:
        print('Progress: '+str(i+1)+' out of '+str(len(short_answers)))
        if type(short_answers[i])==float:
            index.append(-1)
        else:
            a=text[i].find(short_answers[i])
            index.append(a)
    return(index)


short_answers['answer_start']=get_index(df['text'],short_answers['short_answers'])
```

```python
df.head()
```

```python
def is_possible(answer):
    index=[]
    for i in df['Unnamed: 0']:
        print('Progress: '+str(i+1)+' out of '+str(len(answer)))
        if type(answer[i])==float:
            index.append('false')
        else:
            index.append('true')
    return(index)

short_answers['is_impossible']=is_possible(short_answers['short_answers'])
```

```python
short_answers['id']=short_answers['Unnamed: 0']
short_answers['text']=short_answers['short_answers']
```

```python
for i in short_answers['id']:
    short_answers.loc[i,'id']= short_answers.loc[i,'id'].astype(np.float)
    
#print(type(answers.loc[0,'id']))
```

```python
for i in short_answers['id']:
    short_answers.loc[i,'answer_start']= short_answers.loc[i,'answer_start'].astype(np.float)
    
#type(answers.loc[0,'answer_start'])
```

```python
short_answers=short_answers.drop(columns=['Unnamed: 0','short_answers'])
```
